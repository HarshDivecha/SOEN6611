```python

import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import functions
from IPython.display import Image

sns.set_style('darkgrid')
```

### Prepare Data

```python

mockito = functions.process_project_data('mockito',exploratory=True)
wildfly = functions.process_project_data('wildfly',exploratory=True)
ant = functions.process_project_data('ant',exploratory=True)


mockito = functions.limit_data_dict(mockito)
wildfly = functions.limit_data_dict(wildfly)
ant = functions.limit_data_dict(ant)
```

### Picking random version for each system:
- Mockito : v17
- wildfly : v12
- ant : v5 

```python

analysisDict ={
    'Mockito v17':mockito['v17'],
    'Wildfly v12':wildfly['v12'],
    'Ant v5':ant['v5']
}
```

### Boxplots

```python

for ver in analysisDict:
    f,ax = plt.subplots(figsize=(7,5))
    plt.title(ver)
    df = analysisDict[ver]
    sns.boxplot(data=df)
    plt.show()
    
```

### Heatmaps:

```python

functions.plot_heatMaps(df=analysisDict)
```

```python

functions.plot_heatMaps(df=analysisDict,corrThresh=0.5)
```
