```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')

import functions
```

```python
mockito = functions.process_project_data('mockito',exploratory=True)
wildfly = functions.process_project_data('wildfly',exploratory=True)
```

### Exploring outliers :

```python
for ver in mockito:
    print(ver)
    
    f,ax = plt.subplots(figsize=(10,9))
    #ax.set_ylim([0,60])
    df = mockito[ver]
    
    sns.boxplot(data=df)
    plt.show()
```

### Limiting data and reducing outliers

```python
mockito = functions.limit_data_dict(mockito)
wildfly = functions.limit_data_dict(wildfly)
```

```python
for ver in wildfly:
    print(ver)
    
    f,ax = plt.subplots(figsize=(10,9))
    #ax.set_ylim([0,60])
    df = wildfly[ver]
    
    sns.boxplot(data=df)
    plt.show()
```
